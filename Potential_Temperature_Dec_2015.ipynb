```python
import netCDF4
```

```python
# Potential Temperature subset fetched for period (12/01/2015 - 12/31/2015)
nc = netCDF4.Dataset('/content/sample_data/godas_pottmp.2015.nc')
```

```python
# Retrive variable pottmp 
pottmp = nc.variables['pottmp']
```

```python
# pottmp is a 4D array with masked values
pottmp[:]
```

```python
# time is one dimensional which is for December 2015
# fetching data for a single level (first level)
data = pottmp[0][0]
```

```python
data.shape
```

```python
lon = nc.variables['lon']
lat = nc.variables['lat'] 
```

```python
latvals = lat[:]
lonvals = lon[:]
```

```python
latvals
```

```python
lonvals
```

```python
import matplotlib.pyplot as plt
```

```python
#Plotting potential temp data for first level
plt.title (pottmp.long_name + ' (' + pottmp.units + ')')
plt.xlabel(lon.long_name    + ' (' + lon.units    + ')')
plt.ylabel(lat.long_name    + ' (' + lat.units    + ')')
plt.contourf(lonvals, latvals, data, 40, cmap=plt.get_cmap('YlOrRd'))
plt.colorbar()
plt.show()
```

```python
level = nc.variables['level']
levels = level[:]
```

```python
levels
```

```python
print(levels)
```

```python
len(levels)
```

```python
# Visualizing for all levels
for i in range(0,len(levels)):
  data = pottmp[0][i]
  plt.title (pottmp.long_name + ' (' + pottmp.units + ')' + ' for level depth ' + str(levels[i]) + 'm below sea level')
  plt.xlabel(lon.long_name    + ' (' + lon.units    + ')')
  plt.ylabel(lat.long_name    + ' (' + lat.units    + ')')
  plt.contourf(lonvals, latvals, data, 40, cmap=plt.get_cmap('YlOrRd'))
  plt.colorbar()
  plt.show()
```
